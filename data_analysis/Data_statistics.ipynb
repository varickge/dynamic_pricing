# Max count of neighbors  for each Qid1 year-by-year

```python
def max_count(data_path):
    df = pd.read_csv(data_path)
    grouped_df = df.groupby("Qid1")[["Qid2"]].agg(list)
    grouped_df["Qid2_length"] = grouped_df["Qid2"].apply(len)
    return grouped_df["Qid2_length"].max()
```

```python
# for 2019
max_count("./data/distances_2019.csv")
```

```python
# for 2022
max_count("./data/distances_2022.csv")
```

```python
# for 2023
max_count("./data/distances_2023.csv")
```
